```python

directors = []
for roots,dirs,files in os.walk(path):
    directors = dirs
    #just to find the current dirs if i loop on os.walk it will loop on all files at once not one dir at a time
    if ~len(dirs):
        break
        
```

```python
json_files_List = []
for dirc in list_participant_dirs:
    file =  os.path.join(f"{path}/{dirc}/{dirc}_QUESTIONNAIRES.json")
    f = open(file)
    data = json.load(f)
    json_files_List.append(data)
```
